# Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Collection and Analysis

## PIMA Kidney Disease Dataset

```python
# load the dataset
kidney_data = pd.read_csv("/Users/macbookpro/Desktop/ML & DL & GENAI Projects/Machine Learning/ Kidney Disease Predictive System/kidney_disease_dataset.csv")
kidney_data
```

```python
# first 5 rows of the dataset
kidney_data.head()
```

```python
# number of rows of the dataset
kidney_data.shape
```

```python
# get the statistical measures of the dataset
kidney_data.describe()
```

```python
kidney_data['Dialysis_Needed'].value_counts()
```

### 0 ==> No Dialysis
### 1 ==> Dialysis Present

```python
# separating features and the target
X = kidney_data.drop(columns= 'Dialysis_Needed', axis=1)
y = kidney_data['Dialysis_Needed']
```

```python
print(X)
```

```python
print(y)
```

# Data Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

```python
# train the model
classifier =  svm.SVC(kernel='linear')
```

```python
# fit the model
classifier.fit(X_train, y_train)
```

# Model Evaluation

## Accuracy Score

```python
# finding the accuracy score on the training data
X_train_pred = classifier.predict(X_train)
train_data_acc = accuracy_score(X_train_pred, y_train)
```

```python
# print the accuracy score on the training data
print("The Accuracy Score on the Training Data is: ", train_data_acc)
```

```python
# print the accuracy on the testing data
X_test_pred = classifier.predict(X_test)
test_data_acc = accuracy_score(X_test_pred, y_test)
```

```python
# print the accuracy score on the testing data
print("The Accuracy Score on the Testing Score is: ", test_data_acc)
```

# Building the Predictive System

```python
input_data = (71,0.3,40.9,0,1,46.8,1622,1) # No Dialysis Present
# input_data = (28,1.55,14.5,1,0,73.4,1096,0) # Dialysis Present
# change the input data into a numpy array
input_data_as_numpy_array = np.asarray(input_data)
# reshape thr array as we predict for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = classifier.predict(input_data_reshaped)
print(prediction)
# make a predictive approach
if (prediction[0] == 0):
    print("No Dialysis Present")
# elif (prediction[1] == 1):
#     print("Dialysis Present")
else:
    print("Dialysis Present")
```
